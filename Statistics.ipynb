# Concatenate DataFrames

```python
india=pd.DataFrame({
    "city": ["Mumbai","Agra","Banglore"],
    "temp": [34,23,45]
})
india
```

```python
us=pd.DataFrame({
    "city": ["Chicago", "Newyork"," Ireland"],
    "temp": [45,23,67]
})
us
```

```python
df=pd.concat([india,us])
df
```

```python
df=pd.concat([india,us],ignore_index=True)
df
```

```python
df= pd.concat([india,us],axis=1)
df
```

# Merge dataframes

```python
temp=pd.DataFrame({
    "city":["hyd","vzg","N"],
    "tempt":[34,45,67]
})
temp
```

```python
t=pd.DataFrame({
    "city":["hyd","vzg","vijy"],
    "rainfall":[36,55,77]
})
t
```

```python
df=pd.merge(temp,t,on="city")
df
```

```python
df=pd.merge(temp,t,on="city",how="outer")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["rainfall"]=df["rainfall"].fillna(df.rainfall.min())
df
```

```python
df["tempt"]=df["tempt"].fillna(df.tempt.max())
df
```
