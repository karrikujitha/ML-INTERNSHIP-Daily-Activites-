# Binning

```python
Binnig also know as bucketing, is a data preproccessing technique in machine learning used to group continuous numerical variable
into increte intervals or bins
```

```python
each data point is the assigned a label or value corresponding to the bin into which it walls
```

# Why do we use Binning

```python
* Reduce noise in data
* Handle outliner
* simplify complex data
* Improve model interpretability
* Convert continuour--> categorical
```

# Used in

```python
1. decision tree
2. rule-based models
3. business analytics
4. credit scoring , grading system 
```

# Types of Binning

```python
1. Equal-Width Binning
2. Equal-Frequency Binning
3. Custom/Domain-Based Binning
```

# Equal-with Binning

```python
1. it divides the entire range of a continuous variable into a fixed number of bins, where each bin has the same width
2. this method does not consider how data is distributed inside the range
it only focuses on the minimum and maximum values
```

# Formula

```python
Bin width=(max-min)/k
```

# Example

```python
[18,22,25,30,35,40,45,50]
min=18
max=50
k=4
Bin Width= 50-18/4 = 32/4 = 8

18-26=bin1
26-34=bin2
34-42=bin3
42-50=bin4
```

# Key points

```python
1. count different
2. Range same
```

# Code

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "Age":[18,22,25,30,35,40,45,50]
})
print(data)
```

# Equal-Width

```python
data["Age_bin"]=pd.cut(data["Age"],bins=4)
data
```

```python
plt.hist(data["Age"],bins=3)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

```python
plt.scatter(range(len(data)),data["Age"])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Scatter Plot Showing Binning")
plt.show()
```

```python
data["Age_frequency_bin"] = pd.qcut(data["Age"], q=4)
data
```
