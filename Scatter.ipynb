```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,6,8]
y=[2,4,6,7,10]
data=pd.DataFrame({"x":x,"y":y})
print(data)
```

```python
plt.scatter(x,y)
plt.plot(x,y,color="red")
plt.show()
```

```python
X=np.mean(x)
Y=np.mean(y)
print(X,Y)
```

```python
m=np.sum(((X-data.x)*(Y-data.y)))/np.sum((X-x)**2)
print(m)
```

```python
c=Y-(m*X)
print(c)
```

```python
predicition=(m*data.x)+c
print(predicition)
```

```python
plt.scatter(data.x,predicition)
plt.plot(data.x,predicition)
plt.scatter(data.x,data.y)
plt.plot(data.x,data.y)
plt.show()
```
