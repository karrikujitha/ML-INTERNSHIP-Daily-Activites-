```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pandas as pd
```

```python
import pandas as pd
s = pd.Series([1000,2000,3000,5000,9000])
s
```

# Normalization

```python
s.min()
```

```python
s.max()
```

```python
s_norm = (s-s.min())/(s.max()-s.min())
print(s_norm)
```

```python
s_norm = MinMaxScaler().fit_transform(s.values.reshape(-1,1))
print(s_norm)
```

```python
df = pd.DataFrame({
    "temp":[20,25,30,35,40],
    "humid":[30,45,60,70,90]
})
df
```

```python
norm_data = MinMaxScaler().fit_transform(df)
norm_data
```

```python
norm_df = pd.DataFrame(norm_data, columns=df.columns)
norm_data
```

# Standardization

```python
import pandas as pd
s = pd.Series([1000,2000,3000,5000,9000])
s
```

```python
s.mean()
```

```python
s.std()
```

```python
z_Score = (s-s.mean())/s.std()
print(z_Score)
```

```python
z_Score = StandardScaler().fit_transform(s.values.reshape(-1,1))
print(z_Score)
```

```python
z_Score = (s-s.mean())/s.std(ddof=0)
print(z_Score)
```
